```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("ADSexp2dataset.csv")
data.head()
```

```python
data = pd.DataFrame(data.iloc[:,:4])
data
```

# Z-SCORE

```python
import scipy.stats as stats

stats.zscore(data)
```

# Z-TEST

```python
from numpy.random import randn
from statsmodels.stats.weightstats import ztest

x1 = data.loc[:, '5.1']
x2 = data.loc[:, '3.5']
z = ztest(x1, x2, value= 1505, alternative='larger')
z
```

# T-TEST

```python
import scipy.stats as stats
x1 = data.loc[:, '5.1']
x2 = data.loc[:, '3.5']

stats.ttest_ind(x1, x2, equal_var=True)
```

# PAIRED T-TEST

```python
import scipy.stats as stats
x1 = data.loc[:, '5.1']
x2 = data.loc[:, '3.5']

stats.ttest_rel(x1, x2)
```

# CHISQUARE

```python
from scipy.stats import chisquare
x1 = data.loc[:, '5.1']
chisq = chisquare(x1).statistic
pvalue = chisquare(x1).pvalue
print("chisquare statistics: ", chisq)
print("chisquare pvalue: ", pvalue)
```
